# Similarity Functions

```python
def find_relevant_images(txt_ft, images_ft, txt_ft_extra=None, images_ft_extra=None, weight=None, list_images=LIST_IMAGES):
    # if LIST_IMAGES is None --> return index
    # else return value at index in LIST_IMAGES
    sim_score = txt_ft @ images_ft.T
    sim_score = torch.tensor(sim_score)
    if images_ft_extra is not None and txt_ft_extra is not None:
        if weight is None:
            weight = 1
        sim_score = (1-weight)*sim_score + weight * torch.tensor(txt_ft_extra @ images_ft_extra.T)
    value, index = torch.sort(sim_score, descending=True)
    index = index.tolist()
    score = value.tolist()
    results = [list_images[x] for x in index] if list_images is not None else index
    return results, score

def cosine_numpy(A, B):
    # cosine = np.dot(A,B)/(norm(B, axis=1)*norm(A))
    A = A.reshape((1,A.shape[0]))
    cosine = np.sum(A*B, axis=1)/(norm(A, axis=1)*norm(B, axis=1))
    return cosine

def get_hit_rate(pred, gt, hit=[1,5,10]):
    row = []
    for n in hit:
        matches = set.intersection(set(pred[:n]), set(gt))
        if(len(matches)==0):
            row.append(0)
        else:
            row.append(1)
    return row
```

```python
start_idx_list = [0,1,2]
end_idx_list = [1,2,3]
hit_k = [1, 3, 5, 10, 20, 50, 100]

for idx_hint, (start_idx, end_idx) in enumerate(zip(start_idx_list, end_idx_list)):
    res = []
    txt_enc_each = txt_enc[start_idx*num_queries:end_idx*num_queries,:]
    for idx in range(len(txt_enc_each)):
        txt_ft = txt_enc_each[idx]
        gt_relevant = relevant_dict[idx]
        pred_relevant, pred_sim = find_relevant_images(txt_ft, img_enc, list_images=LIST_IMAGES)
        row = get_hit_rate(pred_relevant, gt_relevant, hit=hit_k)
        res.append(row)
    hit = pd.DataFrame(res, columns = ['H@1','H@3','H@5','H@10','H@20','H@50', 'H@100'])
    print(f"===== HINT {idx_hint+1} =====")
    print("1       |3       |5       |10      |20      |50      |100     |")
    print("--------|--------|--------|--------|--------|--------|--------|")
    for value in hit.mean(axis=0).values:
        print(f"{value:0.4f}  |", end='')
    print("")
```
